**Question: Why do you need to perform the train-test split before running a dimensionality reduction algorithm?**

Answer: You need to perform the train-test split before running a dimensionality reduction algorithm because doing so prevents the algortihm from being prematurely fitted onto test data. By splitting the data beforehand and fitting the algorithm to the training data, this allows the algorithm to respond authentically to the test data without knowing anything in advance.

## Part 4: Train classification models on transformed data and original data

In this section, we would like to show that the test accuracy does not decrease too much when we use transformed data. You should do the following:

1. Repeat the train-test split 10 times. For each split:

    - Train a k-nearest neighbor (KNN) model with $k=5$ on the training data (original data without PCA transformation) and report accuracy on test data (original test data without PCA transformation).

    - Implement PCA with target dimensions 1-10. For each transformed data, train another KNN model with $k=5$, and then report the test accuracy. (Note: although you are training 10 new KNN models, you should only do PCA once!)

2. Visualize your results. The x-axis should be the number of principal components (1-10), and the y-axis should be average training accuracy (subplot 1) and test accuracy (subplot 2) over 10 experiments. Standard deviation should also be displayed by adding a shaded region (+/- 1 std. dev.). See sample plot below: labels etc. have been hidden but you should include them!

Your work will be graded based on correctness as well as including informative legends, labels, titles, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

data = load_breast_cancer()
X = data.data
y = data.target
n_features = X.shape[1]
n_splits = 10
n_dims = 10

test_scores = np.zeros((n_splits, n_dims))
train_scores = np.zeros((n_splits, n_dims))
baseline_train = np.zeros(n_splits)
baseline_test = np.zeros(n_splits)

for s in range(n_splits):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    #Baseline KNN (no PCA)
    knn_base = KNeighborsClassifier(n_neighbors=5)
    knn_base.fit(X_train, y_train)
    baseline_train[s] = knn_base.score(X_train, y_train)
    baseline_test[s] = knn_base.score(X_test, y_test)
    #PCA
    pca = Standard_PCA(n_components=n_features)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    for d in range(1, n_dims + 1):
        X_train_pca_d = X_train_pca[:, :d]
        X_test_pca_d = X_test_pca[:, :d]
        knn = KNeighborsClassifier(n_neighbors=5)
        knn.fit(X_train_pca_d, y_train)
        train_scores[s, d-1] = knn.score(X_train_pca_d, y_train)
        test_scores[s, d-1] = knn.score(X_test_pca_d, y_test)


pcs = np.arange(1, n_dims + 1)
train_mean = train_scores.mean(axis=0)
train_std = train_scores.std(axis=0)
test_mean = test_scores.mean(axis=0)
test_std = test_scores.std(axis=0)
base_train_mean = baseline_train.mean()
base_train_std = baseline_train.std()
base_test_mean = baseline_test.mean()
base_test_std = baseline_test.std()

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 4))

#TRAINING average accuracy
ax1.plot(pcs, train_mean, color='cornflowerblue', label='KNN (PCA) Train Average')
ax1.fill_between(pcs, train_mean-train_std, train_mean+train_std, color='cornflowerblue', alpha=0.2)
ax1.axhline(base_train_mean, color='orange', label='KNN (no PCA) Train Average')
ax1.fill_between(pcs, base_train_mean-base_train_std, base_train_mean+base_train_std, color='orange', alpha=0.2)
ax1.set_xlabel('Number of Principal Components')
ax1.set_ylabel('Training Accuracy')
ax1.legend()
ax1.set_title('Average KNN Classifier Training Accuracy w/ PCA vs. No PCA')

#TESTING average accuracy
ax2.plot(pcs, test_mean, color='cornflowerblue', label='KNN (PCA) Test Average')
ax2.fill_between(pcs, test_mean-test_std, test_mean+test_std, color='cornflowerblue', alpha=0.2)
ax2.axhline(base_test_mean, color='orange', label='KNN (no PCA) Test Average')
ax2.fill_between(pcs, base_test_mean-base_test_std, base_test_mean+base_test_std, color='orange', alpha=0.2)
ax2.set_xlabel('Number of Principal Components')
ax2.set_ylabel('Test Accuracy')
ax2.legend()
ax2.set_title('Average KNN Classifier Testing Accuracy w/ PCA vs. No PCA')
plt.show()
```

**Question: How do the results compare between the models trained using the full data vs the PCA components? Why do you think this occurred?**

Answer: The models trained using the PCA components generally performed better compared to the models using the full data. This is likely due to the fact that PCA reduces dimensionality which eliminates noise and redundant features. By ignoring this unnecessary information, the KNN classifier can generalize better and avoid overfitting.
